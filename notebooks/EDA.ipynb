# Análise Exploratória — Vehicles US

Pequena exploração do dataset de anúncios de venda de veículos usados nos EUA.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../vehicles_us.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Histograma — odômetro

```python
fig = px.histogram(df, x="odometer", nbins=50, title="Distribuição do odômetro")
fig.show()
```

## Histograma — preço

```python
fig = px.histogram(df, x="price", nbins=50, title="Distribuição do preço")
fig.show()
```

## Scatter — odômetro vs preço

```python
fig = px.scatter(df, x="odometer", y="price", title="Odômetro x Preço")
fig.show()
```
